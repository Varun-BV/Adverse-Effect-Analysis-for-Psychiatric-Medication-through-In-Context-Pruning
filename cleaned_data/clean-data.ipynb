```python
import re
import json
import hashlib

# Function to generate unique ID based on the username using a hash
def generate_unique_id(username):
    # Generate a unique ID using SHA256 hash of the username
    return hashlib.sha256(username.encode()).hexdigest()

# Function to load data, clean it (remove empty text/comment), and save with unique user IDs
def clean_data(file_path, output_file_path):
    with open(file_path, 'r') as f:
        data = json.load(f)
    
    cleaned_data = []
    user_ids = {}  # To store unique user IDs

    for record in data:
        # Only process if 'text' or 'comment' fields are non-empty
        if 'text' in record and record['text'].strip():
            # Process post with text field
            pass
        elif 'comment' in record and record['comment'].strip():
            # Process post with comment field
            pass
        else:
            # Skip this record if both text and comment are empty
            continue
        
        # Remove the username and replace with unique ID
        username = record.get('author', 'unknown')
        if username not in user_ids:
            user_ids[username] = generate_unique_id(username)
        record['user_id'] = user_ids[username]
        
        # Remove the original author field
        record.pop('author', None)
        
        cleaned_data.append(record)

    # Save the cleaned data to a new JSON file
    with open(output_file_path, 'w') as f:
        json.dump(cleaned_data, f, indent=4)

    print(f"Cleaned data saved to {output_file_path}")

# Optional: If you want to run the script independently to clean and save data
if __name__ == "__main__":
    cleaned_data_file = "cleaned_insomnia.json"
    clean_data("/kaggle/input/scraped/scrape_insomnia.json", cleaned_data_file)
```
